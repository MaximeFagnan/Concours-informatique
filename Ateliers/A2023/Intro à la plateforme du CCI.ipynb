# Introduction aux concours informatique avec CCI

## Pourquoi un concours informatique?

Les concours, particulièrement en informatique, sont une manière ludique de développé vos compétences. Les tâches sont stimulantes, à votre portée et vous permettent de vous mettre de nouveaux buts de manières périodiques.

**Exemple 1:**

*Je sais déjà programmer, mais j'aimerais apprendre à programmer en Python. Je me met comme défi de répondre aux questions de niveau 1 à 4 dans le concours de niveau Junior en février. Je me sers des concours passé ainsi que leurs correcteurs pour apprendre et je met au défi mon apprentissage dans l'environnement ludique du concours au mois de février.*

**Exemple 2:**

*J'ai eu un score de 67/75 au concours junior de l'année passée. J'aimerais en apprendre plus sur les algorithmes, ainsi que leurs implémentations en contexte. J'aimerais avoir au moins 25/75 au concours sénior cette année (en 2023, 25/75 te plaçait dans le top 20%). Pour étudier, j'aimerais compléter les 3 premières questions des concours sénior pour les 2 premières années d'ici janvier et être capable de compléter les 3 premières questions du concours sénior en 1 journée en février pour me pratiquer.*

## Description du CCI

Le Concours Canadien d'Informatique ([CCI](https://cemc.uwaterloo.ca/contests/computing/details-f.html)) est organisé tous les ans par l'université de Waterloo.

> *Le CCI a lieu à l’école en utilisant le correcteur informatisé. Les élèves ont 3 heures pour répondre à 5 questions, tant au niveau intermédiaire que supérieur. Les enseignants doivent inscrire leurs étudiants pour pouvoir faire le concours dans leur école. Il n’y a aucune compétition par équipe.*

>Les élèves participant au CCI peuvent utiliser les langages C, C++, Python 2, Python 3 ou Java. Consultez les [exemples de programmes et les détails sur ces langages](https://cccgrader.com/sample_soln.pdf).

## Format du CCI

Le CCI est formé de deux concours, soit le concours de niveau intermédiaire et le concours de niveau supérieur. Chaque concours est composé de 5 questions. Chaque question vaut 15 points. Les questions sont généralement présentées en ordre croissant de difficulté de la première question à la dernière dans chaque concours. 

## Comment se pratiquer ?

1. Vous pouvez trouver les problèmes de votre choix au [lien](https://cemc.uwaterloo.ca/contests/past_contests.html#ccc) suivant. Aucun des problèmes n'a de solutionnaires détaillés, mais les indices sont excellents lorsque vous êtes coincés.

2. Vous pouvez également trouver tous les problèmes directement sur la plateforme du correcteur automatisé https://cccgrader.com/.

3. Vous créer un programme dans le langage de votre choix pour répondre à la question. 
*Je vous conseille d'utiliser la même configuration de travail que le jour du concours (même éditeur de code et même filière pour votre fichier)*.

4. Vous soumettez votre programme dans le correcteur automatisé et vous rafraîchissez la page après quelques secondes pour voir votre résultat.
