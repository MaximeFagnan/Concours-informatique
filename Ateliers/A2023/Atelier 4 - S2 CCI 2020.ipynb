#### Structure du parcours en largeur

1. Vous devez créer deux listes (ou ensembles).
    1. L'ensemble des sommets visité (initialisé vide).
    2. La queue des sommets qui sont a visité dans l'ordre (initialisé avec le ou les sommets de départs).
    
2. Tant qu'il reste des sommets a visités:
    1. Visité celui au dessus de la queue (x) et enlevé le de la queue. Avez-vous fini votre parcours?
    2. Marqués (x) comme sommet visité.
    3. Rajouté les sommets qui sont adjacents à (x) et qui ne sont pas visités.

3. Vous avez fini parcouru tout les sommets connexes au sommet de départ.
    
   

```python
# Exemple en code:
graph = {
  'A' : ['B','C'],
  'B' : ['D', 'E'],
  'C' : ['F'],
  'D' : [],
  'E' : ['F'],
  'F' : []
}

visited = [] #Liste pour conserver les éléments visités.
queue = []     #Initialisons la queue

def bfs(visited, graph, node):
  visited.append(node)
  queue.append(node)

  while queue:
    s = queue.pop(0) 
    print (s, end = " ") 

    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)

# Driver Code
bfs(visited, graph, 'A')
```
